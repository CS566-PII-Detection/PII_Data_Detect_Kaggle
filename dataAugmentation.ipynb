```python
from faker import Faker
from  tqdm.auto import tqdm
import json
import random
fake = Faker()
```

```python
def generate_fake_social_media_urls(num_urls=1):
    social_media_platforms = {
        'LinkedIn': 'linkedin.com/in/',
        'YouTube': 'youtube.com/c/',
        'Instagram': 'instagram.com/',
        'GitHub': 'github.com/',
        'Facebook': 'facebook.com/',
        'Twitter': 'twitter.com/'
    }

    fake_social_media_urls = []

    for _ in range(num_urls):
        fake_user_name = fake.user_name()
        platform, domain = random.choice(list(social_media_platforms.items()))
        fake_url = f'https://{domain}{fake_user_name}'
        fake_social_media_urls.append(fake_url)

    return fake_social_media_urls[0]
```

```python
NUM = 2000
fake_identities = []
Faker.seed(1)
for i in tqdm(range(NUM)):
    temp_d = {
        "NAME_STUDENT":[],
        "EMAIL":[],
        "USERNAME":[],
        "ID_NUM":[],
        "PHONE_NUM":[],
        "URL_PERSONAL":[],
        "STREET_ADDRESS":[]
    }
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        name = random.choices([fake.name,fake.first_name, fake.last_name], weights = [0.6,0.15,0.15], k = 1)[0]()
        temp_d['NAME_STUDENT'].append(name)
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        temp_d['EMAIL'].append(fake.ascii_free_email())
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        temp_d['USERNAME'].append(fake.user_name())
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        name = random.choices([fake.ssn,fake.passport_number, fake.bban, fake.iban, fake.license_plate], weights = [0.20,0.20,0.20,0.20, 0.20], k = 1)[0]()
        temp_d['ID_NUM'].append(name)
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        temp_d['PHONE_NUM'].append(fake.phone_number())
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        temp_d['URL_PERSONAL'].append(generate_fake_social_media_urls(1))
    for j in range(random.choices([0,1,2,3,4], weights=[0.05,0.7,0.2,0.025,0.025], k=1)[0]):
        temp_d['STREET_ADDRESS'].append(fake.address())
    fake_identities.append(temp_d)
```

```python
json.dump(fake_identities, open('fake_data.json', 'w'))
```

```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation', model='lmsys/vicuna-7b-v1.5', device=0)
```

```python
def create_prompt(info):
    # prompt = "You are an essay writer. Given information includes:"
    # prompt = "You are an essay writer. You will be given some personal information about the student writing it, like name email etc as a dictionary. You have to write an essay that includes all the given information somewhere in the essay do not miss out any. Given information has the following keys: NAME_STUDENT - The full or partial name of a student that is not necessarily the author of the essay. This excludes instructors, authors, and other person names. EMAIL - A student’s email address. USERNAME - A student's username on any platform. ID_NUM - A number or sequence of characters that could be used to identify a student, such as a student ID or a social security number. PHONE_NUM - A phone number associated with a student. URL_PERSONAL - A URL that might be used to identify a student. STREET_ADDRESS - A full or partial street address that is associated with the student, such as their home address. Each of these will have a list of values associated with them. Ensure that you include each and every one of them in the essay. do not miss out any."
    prompt = "You are an essay writer. You will be given some personal information about the student writing it, like name email etc as a dictionary. You have to write an essay that includes all the given information somewhere in the essay do not miss out any."
    prompt += 'Given information includes:\n'
    for key, values in info.items():
        prompt += f" {key} - {', '.join(values)}.\n"
    prompt += " Write an essay that includes all the given information."
    return prompt
```

```python
def find_indices(lst, element):
    return [index for index, value in enumerate(lst) if value == element]
```

```python
def convert_to_tokenlevel(para, label):
    # labs = eval(label)
    tok = para.split()
    labb = ['O']*len(tok)
    ws = [True]*len(tok)
    ws[-1] = False
    for i in (label):
        for j in label[i]:
            k = j.split()
            b_flag = True
            for m in k:
                if b_flag:
                    indices = find_indices(tok,m)
                    for ind in indices:
                        labb[ind] = 'B-'+i
                    b_flag = False
                else:
                    indices = find_indices(tok,m)
                    for ind in indices:
                        labb[ind] = 'I-'+i
    return {
        "tokens":tok,
        "labels":labb,
        "trailing_whitespace":ws
    }
```

```python
more = []
```

```python
batch_size = 5
ps = []
ps_len = []
labels = []

for i in range(len(fake_identities)//batch_size):
    p = []
    p_len = []
    ll = []
    for j in range(batch_size):
        pp = create_prompt(fake_identities[i*batch_size+j])
        p.append(pp)
        p_len.append(len(pp))
        ll.append(fake_identities[i*batch_size+j])
    ps.append(p)
    ps_len.append(p_len)
    labels.append(ll)
```

```python
for p, p_len, label in zip(ps, ps_len, labels):
    essay = generator(p, max_length=5000, num_return_sequences=1,temperature=0.5,repetition_penalty=1.01)
    for e,l,ll in zip(essay,p_len,label):
        more.append(convert_to_tokenlevel(e[0]['generated_text'][l:],ll))
```
